```python
#import packages/library
import pandas as pd
import sklearn
from sklearn.cluster import KMeans
```

```python
#import KMeans
from sklearn.cluster import KMeans
```

```python
#panggil dan tampilkan dataset diabetes.csv
dataset_Diabetes = pd.read_csv('diabetes.csv')
dataset_Diabetes.head()
```

```python
#melihat tipe data pada dataset
dataset_Diabetes.info()
```

```python
#membuat objek dengan algoritma KMeans dengan jumlah cluster 4
kmDiabet = KMeans(n_clusters=4)
```

```python
#pilih atribut yang akan dilatih -> glukosa dan tekanan darah
pointDiabet = dataset_Diabetes[["Glucose","BloodPressure"]]
```

```python
kmDiabet.fit(pointDiabet)
```

```python
dataset_Diabetes['cluster'] = kmDiabet.labels_
```

```python
dataset_Diabetes.head()
```

```python
labelDiabet = kmDiabet.labels_
```

```python
labelDiabet
```

```python
#membuat dataframe dengan kolom label dan Age  
dfDiabet = pd.DataFrame({'label': labelDiabet, 'BloodPressure': dataset_Diabetes.cluster})
```

```python
#membuat crosstab
crosstabDiabet = pd.crosstab(dfDiabet["label"],dfDiabet["BloodPressure"])
```

```python
#menampilkan crosstabDiabet
crosstabDiabet
```

```python
#menampilkan pusat cluster=>centroid
centroid = kmDiabet.cluster_centers_
centroid
```

#### Visualisasi 

```python
X = pointDiabet['Glucose']
Y = pointDiabet['BloodPressure']
```

```python
# import tools untuk visualisasi
import matplotlib.pyplot as plt

# Menampung koordinat dari tiap centroid
centroid_x = centroid[:,0]
centroid_y = centroid[:,1]
```

```python
# visualisasi
plt.figure(figsize=(5,5))
plt.scatter(X, Y, alpha=0.5, c=labelDiabet)
plt.scatter(centroid_x, centroid_y, marker="D", s=50,label="Centroid")
plt.title("Hasil visualisasi klustering KMeans")
plt.xlabel("Konsentrasi glukosa dlm test(mm)")
plt.ylabel("Tekanan darah(mmHg)")
plt.legend()
plt.show()
```

#### Jumlah cluster

Elbow Curve : merupakan salah satu metode yang bisa digunakan untuk menemukan jumlah optimal dari cluster (k)

```python
jarak_cluster = range(1, 6)
inertias = []

for k in jarak_cluster:
    kmDiabet = KMeans(n_clusters=k)
    kmDiabet.fit(pointDiabet)
    inertias.append(kmDiabet.inertia_)
    
# Plot jarak cluster dengan inertias
plt.figure(figsize=(10,10))
plt.plot(jarak_cluster, inertias, '-o')
plt.xlabel('Cluster ke')
plt.ylabel('inertia/SSE')
plt.xticks(jarak_cluster)
plt.show()
```
