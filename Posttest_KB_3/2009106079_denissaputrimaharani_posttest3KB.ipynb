```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_rental=pd.read_csv("CarRentalData.csv")
data_rental
```

# Grafik batang

```python
x = data_rental['vehicle.type']
y = data_rental['rate.daily']
```

```python

plt.bar(x, y, color='rosybrown', width=.5, align='center')
plt.title('Perbandingan tarif sewa harian antar kendaraan')
plt.xlabel('Tipe kendaraan')
plt.ylabel('Tarif sewa harian')
plt.show()
```

kesimpulan : tipe kendaraan car (mobil keluarga) adalah tipe kendaraan yang paling banyak di sewa di US dan tipe kendaraan yang sedikit peminatnya adalah minivan dengan tarif sewa harian dibawah 200.

# Grafik PIE 

```python
data_rental['fuelType'].value_counts()
```

```python
bahanBakar = data_rental['fuelType'].value_counts() .rename_axis('Jenis bahan bakar') .reset_index(name='jumlah rating')
rating = data_rental['rating'].value_counts()
```

```python
bahanBakar.head()
```

```python
fig = plt.figure(figsize=(10,10), dpi=100)
colors=('#C9CBA3', '#FFE1A8', '#E26D5C', '#723D46')
plt.pie(data_rental['fuelType'].value_counts(), labels = bahanBakar['Jenis bahan bakar'], autopct='%.1f%%', colors=colors, startangle=100 )
plt.title('Presentase rating kendaraan berdasarkan jenis bahan bakar')
plt.show()
```

kesimpulan : masyarakat amerika lebih memilih menggunakan kendaraan berbahan bakar bensin dibanding diesel

# Scatter plot

```python
data_kendaraan = pd.read_csv('CarRentalData.csv')
data_kendaraan.head(10)
```

```python
sns.scatterplot(data = data_kendaraan, x = 'vehicle.year', y = 'rating')
plt.title('data rating kendaraan')
plt.xlabel('tahun kendaraan')
plt.ylabel('rating')
plt.show()
```

kesimpulan : rata rata kendaraan yang dibuat tahun 2010-2020 memiliki rating diatas 3 yang berarti semakin lama tahun kendaraan dibuat semakin sedikit yg menyewa dan grafik diatas menggambarkan bahwa tahun pembuatan kendaraan juga menjadi pertimbangan masyarakat ketika ingin menyewa kendaraan.
